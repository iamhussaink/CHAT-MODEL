```python
from api_key import myapi #another .py file made by me to hide my api_key
api_key = myapi()
```

```python
from langchain.llms import GooglePalm
llm = GooglePalm(google_api_key=api_key, temperature=0.7)
```

```python
from langchain.document_loaders import UnstructuredURLLoader
```

```python
loader = UnstructuredURLLoader(urls=["https://en.wikipedia.org/wiki/Earth"])
```

```python
data =loader.load()
```

```python
data[0].page_content[0:100]
```

```python
data[0].metadata
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
recur = RecursiveCharacterTextSplitter(separators=['.','\n'], chunk_size=250, chunk_overlap=20 )
```

```python
chunks = recur.split_documents(data)
```

```python
len(chunks)
```

```python
chunks[347]
```

```python
from langchain_community.vectorstores import FAISS
```

```python
from langchain.embeddings import GooglePalmEmbeddings
```

```python
embeddings = GooglePalmEmbeddings(google_api_key=api_key)
```

```python
vector = FAISS.from_documents(chunks,embeddings)
```

```python
from langchain.chains import RetrievalQA
```

```python
chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vector.as_retriever())
```

```python
query = "what is the radius of earth"
```

```python
result = chain.invoke(query)
```
